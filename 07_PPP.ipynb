```python
from cobra.io import read_sbml_model, write_sbml_model
from cobra import Reaction, Metabolite
import matplotlib.pyplot as plt
from cobra.flux_analysis import production_envelope
```

```python
model = read_sbml_model('data/iML1515_het.xml')
```

## Constructing phenotype phase plane plot for P3HB:

```python
# PPP:
with model:
    r_phaC_exp = model.add_boundary(model.metabolites.get_by_id('P3HB_c'), type='sink')
    ppp_p3HB = production_envelope(model,
                        reactions=[model.reactions.get_by_id('SK_P3HB_c')],
                        objective=model.reactions.get_by_id("BIOMASS_Ec_iML1515_core_75p37M"))
    
pppp = ppp_p3HB.plot(x='SK_P3HB_c', y='flux_maximum')
```
